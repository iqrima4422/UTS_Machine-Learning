```python
# Library Python
import numpy as np
import pandas as pd

df = pd.read_csv('tweet_emotions_tokenize.csv', encoding='latin-1') # spesifikasi encoding diperlukan karena data tidak menggunakan UTF-8

df.head()
```

```python
# Data untuk rename kolom v1 dan v2
new_cols = {
    'content' : 'Content_tweet',

}
# Rename nama kolom v1 dan v2
df = df.rename(columns=new_cols)

# cek data
df.head()
```

```python
#INSPEKSI DATA 

# Cek Jumlah Data Per Kelas
print (df['sentiment'].value_counts())
print('\n')

# Cek Kelengkapan Data
print(df.info())
print('\n')

# Cek Statistik Deskriptif
print (df.describe())
```

```python
# Encode Label

# Data untuk label 
new_labels = {
    'neutral' : 0,
    'worry' :1,
    'happiness' :2,
    'sadness' :3,
    'love' :4,
    'surprise' :5,
    'fun' :6,
    'relief' :7,
    'hate' :8,
    'empty' :9,
    'enthusiasm' :10,
    'boredom' :11,
    'anger' :12,
}

# Encode label
df['sentiment'] = df['sentiment'].map(new_labels)

# Cek data
df.head()
```

```python
# Memisahkan Fitur dengan Label

X = df['Content_tweet'].values.astype('U')
y = df['sentiment'].values.astype('U')
```

```python
# Extraksi Fitur

from cgi import test
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

#Split data training dan testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)

#Inisiasi TfidfVectorizer
bow = TfidfVectorizer(stop_words='english')


#Fitting dan transform X_train dengan TF-Idf
X_train = bow.fit_transform(X_train)

#Transform X_test
#Mengapa hanya transform? Alasan yang sama dengan kasus pada percobaan ke-3
#Kita tidak menginginkan model mengenai parameter yang digunakan oleh CountVektor


#  menskalakan atau menstandardisasi data training dan testing
X_test = bow.transform(X_test)
print(f'Hasil standarisasi data : {X_test}')
```

```python
# Training dan Evaluasi Model

from numpy import average
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report , precision_score

# Inisiasi MultinomialNB
mnb = MultinomialNB()

# Fit model
mnb.fit(X_train, y_train)

# Prediksi dengan data training
y_pred_train = mnb.predict(X_train)

# Evaluasi akurasi data training
acc_train = accuracy_score(y_train, y_pred_train)

# Prediksi dengan data training
y_pred_test = mnb.predict(X_test)

# Evaluasi akurasi data training
acc_test = accuracy_score(y_test, y_pred_test)

# Menghitung Precission
pre_test = precision_score (y_test, y_pred_test, average='weighted')

print(classification_report(y_test, y_pred_test))

# Print hasil evaluasi
print(f'Hasil akurasi data train: {acc_train}')
print(f'Hasil akurasi data test: {acc_test}')
print(f'Hasil Presisi data test: {pre_test}')
```
